<h5>K-Means Clustering</h5>

<p>Clustering Dummy Data with K-Means Clustering</p>

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets._samples_generator import make_blobs

sns.set_style("darkgrid")
plt.rcParams["figure.figsize"] = (10, 8)
```

```python
# Generating dummy data of 500 records with 4 clusters
features, labels = make_blobs(n_samples=500, centers=4, cluster_std=2.0, random_state=42)

# Plotting the dummy data
plt.scatter(features[:, 0], features[:, 1])
```

```python
# Performing kmeans clustering using KMeans class
km_model = KMeans(n_clusters=4)
km_model.fit(features) 
```

```python
# Printing the centroid values
print(km_model.cluster_centers_)
```

```python
# Printing the predicted label values
print(km_model.labels_)
```

```python
# Print the data points
plt.scatter(features[:, 0], features[:, 1], c=km_model.labels_, cmap='rainbow')

# Print the centroids
plt.scatter(km_model.cluster_centers_[:, 0], km_model.cluster_centers_[:, 1], s=100, c='black')
```

<p>Print the actual data points</p>

```python
plt.scatter(features[:, 0], features[:, 1], c=labels, cmap='rainbow')
```

<p>Customer Segmentation using KMeans clustering</p>

```python
dataset = pd.read_csv('./datasets/mall_customers.csv')
dataset.head()
```

```python
dataset.shape
```

```python
# Plotting the histogram for the annual income column
sns.distplot(dataset['Annual Income (k$)'], kde=False, bins=50)
```

```python
# Plotting the histogram for the spending score column in red color with 50 bins
import seaborn as sns
sns.distplot(dataset['Spending Score (1-100)'], kde=False, bins=50, color='red')
```

```python
# Plotting the regression plot for annual income against spending score
sns.regplot(x='Annual Income (k$)', y="Spending Score (1-100)", data=dataset)
```

```python
# Plotting regression plot for age against spending score
sns.regplot(x='Age', y='Spending Score (1-100)', data=dataset, color='green')
```

```python
# Filtering the annual income and spending score columns and removing the remaining columns
dataset = dataset.filter(['Annual Income (k$)', 'Spending Score (1-100)'], axis=1)

# Printing the first 5 rows of the filtered dataset
dataset.head()
```

```python
# Performing kmeans clustering using the KMeans class
km_model = KMeans(n_clusters=4)
km_model.fit(dataset)
```

```python
# Printing centroid values
print(km_model.cluster_centers_)
```

```python
# Printing the predicted label values
print(km_model.labels_)
```

```python
# Print the data points
plt.scatter(dataset.values[:, 0], dataset.values[:, 1], c=km_model.labels_, cmap='rainbow')

# Print the centroids
plt.scatter(km_model.cluster_centers_[:, 0], km_model.cluster_centers_[:, 1], s=100, c='black')
```

```python
# Training kmeans on K Values from 1 to 10
loss = []

for i in range(1, 11):
    km = KMeans(n_clusters=i).fit(dataset)
    loss.append(km.inertia_)

# Printing loss against number of clusters
plt.plot(range(1, 11), loss)
plt.title("Finding Optimal Clusters via Elbow Method")
plt.xlabel("Number Of Clusters")
plt.ylabel("Loss")
plt.show()
```

```python
# Performing kmeans clustering using KMeans class
km_model = KMeans(n_clusters=5)
km_model.fit(dataset)
```

```python
# Print the data points
plt.scatter(dataset.values[:, 0], dataset.values[:, 1], c=km_model.labels_, cmap='rainbow')

# Print the centroids
plt.scatter(km_model.cluster_centers_[:, 0], km_model.cluster_centers_[:, 1], s=100, c='black')
```

<h5>Hierarchical Clustering</h5>

<p>Clustering Dummy Data</p>

```python
# Generating dummy data of 10 records with 2 clusters
features, labels = make_blobs(n_samples=10, centers=2, cluster_std=2.00)

# Plotting the dummy data
plt.scatter(features[:, 0], features[:, 1], color='r')

# Adding numbers to data points
annots = range(1, 11)
for label, x, y in zip(annots, features[:, 0], features[:, 1]):
    plt.annotate(label, xy=(x, y), xytext=(-3, 3), textcoords='offset points', ha='right', va='bottom')
    plt.show()
```

```python
# Importing dendogram and linkage classes from scipy
from scipy.cluster.hierarchy import dendrogram, linkage

# Creating an object of linkage classes
dendos = linkage(features, 'single')

annots = range(1, 11)

# Creating dendrograms
dendrogram(
    dendos,
    orientation='top',
    labels=annots,
    distance_sort='descending',
    show_leaf_counts=True
)

# Printing dendrograms
plt.show()
```

<p>Agglomerative Clustering</p>

```python
from sklearn.cluster import AgglomerativeClustering

# Training agglomerative clustering model
hc_model = AgglomerativeClustering(n_clusters=2, linkage='ward')
hc_model.fit_predict(features)
```

```python
# Print the data points
plt.scatter(features[:, 0], features[:, 1], c=hc_model.labels_, cmap='rainbow')
```

<p>Clustering the Iris Dataset</p>

```python
iris_df = sns.load_dataset('iris')
iris_df.head()
```

```python
# Dividing data into features and labels
features = iris_df.drop(['species'], axis=1)
labels = iris_df.filter(['species'], axis=1)
features.head()
```

```python
# Training hierarchical clustering model
from sklearn.cluster import AgglomerativeClustering

# Training Agglomerative Clustering model 
features = features.values
hc_model = AgglomerativeClustering(n_clusters=3, linkage='ward')
hc_model.fit_predict(features)
```

```python
# Print the predicted cluster data points
plt.scatter(features[:, 0], features[:, 1], c=hc_model.labels_, cmap='rainbow')
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 8))
plt.title('Iris Dendrogram')
dend = shc.dendrogram(shc.linkage(features, method='ward'))
```
